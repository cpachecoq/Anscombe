# Anscombe Practica

## Dataset Asncombe

```python
import pandas as pd

# Especifica la ruta de tu archivo CSV
archivo_csv = 'df_anscombe.csv'

# Carga el archivo CSV en un DataFrame de pandas
df_anscombe = pd.read_csv(archivo_csv)

# Muestra los registros del DataFrame para verificar que se haya cargado correctamente
df_anscombe
```

## Descripcion de los conjuntos del dataset

```python
#Describimos el dataset
df_anscombe.groupby('group').describe()
```

## Grafica de los 4 conjutos

```python
import matplotlib.pyplot as plt

# Dividir el DataFrame en cuatro conjuntos de datos diferentes
conjunto1 = df_anscombe[df_anscombe['group'] == 1]
conjunto2 = df_anscombe[df_anscombe['group'] == 2]
conjunto3 = df_anscombe[df_anscombe['group'] == 3]
conjunto4 = df_anscombe[df_anscombe['group'] == 4]

# Crear una figura con cuatro subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Graficar cada conjunto de datos en su respectivo subplot
axs[0, 0].scatter(conjunto1['x'], conjunto1['y'])
axs[0, 0].set_title('Conjunto I')

axs[0, 1].scatter(conjunto2['x'], conjunto2['y'])
axs[0, 1].set_title('Conjunto II')

axs[1, 0].scatter(conjunto3['x'], conjunto3['y'])
axs[1, 0].set_title('Conjunto III')

axs[1, 1].scatter(conjunto4['x'], conjunto4['y'])
axs[1, 1].set_title('Conjunto IV')

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
